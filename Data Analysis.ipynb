```python
# ðŸ“Š Data Analysis and Visualization on DIA_testset_RDKit_descriptors.csv

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# -----------------------------
# Task 1: Load and Explore the Dataset
# -----------------------------
try:
    df = pd.read_csv("DIA_testset_RDKit_descriptors.csv")
    print("âœ… File loaded successfully!")
except FileNotFoundError:
    print("âŒ Error: File not found. Please check the file name and path.")

# Display first 5 rows
print("\nðŸ”¹ First 5 Rows of the Dataset:")
print(df.head())

# Show structure and missing values
print("\nðŸ”¹ Dataset Info:")
print(df.info())

print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

# Drop rows with missing values
df.dropna(inplace=True)
print("\nâœ… Missing values removed. Shape of cleaned dataset:", df.shape)

# -----------------------------
# Task 2: Basic Data Analysis
# -----------------------------
print("\nðŸ“ˆ Descriptive Statistics:")
print(df.describe())

print("\nðŸ“Œ Median Values:")
print(df.median(numeric_only=True))

print("\nðŸ“Œ Standard Deviation:")
print(df.std(numeric_only=True))

# -----------------------------
# Task 3: Data Visualization
# -----------------------------

# Add index column for plotting trends
df["Index"] = range(len(df))

# 1. Line Chart 
plt.figure(figsize=(8, 4))
sns.lineplot(x="Index", y="MolWt", data=df)
plt.title("Molecular Weight Trend Over Index")
plt.xlabel("Index")
plt.ylabel("MolWt")
plt.show()

# 2. Bar Chart
if 'Label' in df.columns:
    plt.figure(figsize=(8, 4))
    sns.barplot(x="Label", y="MolWt", data=df)
    plt.title("Average Molecular Weight by Label")
    plt.xlabel("Category")
    plt.ylabel("MolWt")
    plt.show()
else:
    print("âš ï¸ 'Label' column not found. Skipping bar chart.") 

# 3. Histogram 
plt.figure(figsize=(6, 4))
sns.histplot(df["MolWt"], bins=20, kde=True)
plt.title("Distribution of Molecular Weight")
plt.xlabel("MolWt")
plt.ylabel("Frequency")
plt.show()

# 4. Scatter Plot (MolWt vs TPSA)
plt.figure(figsize=(6, 4))
sns.scatterplot(x="MolWt", y="TPSA", data=df)
plt.title("MolWt vs TPSA")
plt.xlabel("Molecular Weight")
plt.ylabel("Topological Polar Surface Area (TPSA)")
plt.show()
```
